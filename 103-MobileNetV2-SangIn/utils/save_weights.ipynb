```python
import os
import torch
import torchvision.models as models

from safetensors.torch import save_file
```

### Save weights as safetensors

```python
# save weight as safetensors
SAVE_PATH = "../weights"
if not os.path.exists(SAVE_PATH):
    os.makedirs(SAVE_PATH)

mobilenet_v2 = models.mobilenet_v2(weights=models.MobileNet_V2_Weights.IMAGENET1K_V1)
save_file(mobilenet_v2.state_dict(), os.path.join(SAVE_PATH, "mobilenet_v2_imagenet1k.safetensors"))
```
